# Time Series Data:
#### A time series is a set of observations on the values that a variable takes at different times. Time Series is used in statistics, econometrics, mathemetical finance, weather forecasting, earthquake prediction and many other applications.

## Univariate time series:
#### A time series that consists of single obervations recorded at regular time intervals.

## Cross Sectional Data:
#### Consists of number of variables/ observations at a given point of time
